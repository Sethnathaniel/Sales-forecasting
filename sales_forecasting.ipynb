## Objective
The goal of this project is to analyze historical sales data and build a simple model to forecast future sales.

This project focuses on:
- Understanding sales trends over time
- Visualizing patterns in the data
- Building a basic forecasting model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/sales.csv")
df.head(19)
```

```python
df.info()
```

## Data Overview
This dataset contains historical sales data with dates and corresponding sales values.


```python
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
```

```python
plt.figure()
plt.plot(df["date"], df["sales"])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Over Time")
plt.show()
```

## Sales Trend Analysis
The plot shows how sales evolve over time.
We can observe overall trends and potential seasonal patterns.
This visualization helps understand the behavior of sales before modeling.

```python
df["time_index"] = np.arange(len(df))
df.head(18)
```

```python
#!pip install scikit-learn #to be uncommented if want to install scikit-learn

from sklearn.linear_model import LinearRegression

X = df[["time_index"]]
y = df["sales"]

model = LinearRegression()
model.fit(X, y)

df["sales_pred"] = model.predict(X)
```

```python
# Sales over time
plt.figure()
plt.plot(df["date"], df["sales"], label="Actual Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Over Time")
plt.savefig("visuals/sales_over_time.png", dpi=300)
plt.show()

# Actual vs predicted
plt.figure()
plt.plot(df["date"], df["sales"], label="Actual")
plt.plot(df["date"], df["sales_pred"], label="Predicted")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Actual vs Predicted Sales")
plt.legend()
plt.savefig("visuals/actual_vs_predicted.png", dpi=300)
plt.show()

```

## Model Interpretation
The linear regression model captures the general sales trend over time.
While it is simple, it provides a baseline for sales forecasting.
More advanced models could improve accuracy by accounting for seasonality and external factors.
